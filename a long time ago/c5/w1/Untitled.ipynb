```python
import pandas as pd
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
milk = pd.read_csv('monthly-milk-production.csv',';', index_col=['month'], parse_dates=['month'], dayfirst=True)
```

```python
plt.plot(milk)
```

```python
import statsmodels.api as sm
pval = sm.tsa.stattools.adfuller(milk['milk'])[1]
round(pval, 2)
```
